```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
# Countplot : Survival based as gender
sns.countplot(df, x ='survived',hue='sex', palette='RdBu')
plt.title('Survived vs nplot survived by gender')
plt.show()
```

```python
sns.barplot(df, x='class', y='fare' ,hue='class')
plt.title('Average Fare by passenger class and Gender')
plt.show()
```

```python
sns.violinplot(data=df, x='class', y='age',hue='sex' )
plt.title('Average age ')
plt.show()
```

```python
sns.pointplot(df , x='class',y='survived', hue='sex', palette='dark', markers=['o','x'])
plt.title('Survival Rate by class and Gender')
plt.show
```
